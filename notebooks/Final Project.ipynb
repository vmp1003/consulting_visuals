```python
import pandas as pd
import data_work

import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
```

```python
df = data_work.load_data()
df.head()
```

```python
df = data_work.clean_values(df)
df.head()
```

```python
df = data_work.summary_calculations(df)
df.head()
```

```python
df_detail = data_work.detail_calculations(df)
df_detail.head()
```

```python
df_detail.tail()
```

## Amount Due By State
This graph shows the amount due depending on the  top 10 states. We chose to show this graph because we thought it was interesting for Texas to have a higher amount than California. 

```python
df_detail.groupby('state', as_index=False).agg({'amt_due':'sum'}).sort_values('amt_due')[-10:].plot.bar(y = 'amt_due', x= 'state')
```

## Amount Due By City 
This graph shows the amount due depending on the city. As you can see Washington and Portland have the greatest amount due, and Phoenix having the least amount due. 

```python
df_detail.groupby('city', as_index=False).agg({'amt_due':'sum'}).sort_values('amt_due')[-10:].plot.bar(y = 'amt_due', x= 'city')
```

## Amount Due By Industry
After looking at the amount due by state and city, we wanted to take a look at the top 15 industries. Major Banks, Major Pharmaceuticals have the highest amount due compared to any other industry. 

```python
df_detail.groupby('industry', as_index=False).agg({'amt_due':'sum'}).sort_values('amt_due')[-15:].plot.bar(y = 'amt_due', x= 'industry')
```

## Amount Due by Industry 
This graph is the same as above as a pie chart, giving another way to look at how the industries are broken up by their amount due. 

```python
df_detail.groupby('industry', as_index=False).agg({'amt_due':'sum'}).sort_values('amt_due')[-15:].plot.pie(y = 'amt_due', x= 'industry')
```

```python
df_v = df_detail.reset_index()
df_v['year'] = df_v['due_date'].dt.year
df_v['month'] = df_v['due_date'].dt.month
df_v.head()
```
