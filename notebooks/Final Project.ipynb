```python
import pandas as pd
import data_work

import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
```

```python
df = data_work.load_data()
df.head()
```

```python
df = data_work.clean_values(df)
df.head()
```

```python
df = data_work.summary_calculations(df)
df.head()
```

```python
df_detail = data_work.detail_calculations(df)
df_detail.head()
```

```python
df_detail.tail()
```

```python
df_detail.groupby('state', as_index=False).agg({'amt_due':'sum'}).sort_values('amt_due')[-10:].plot.bar(y = 'amt_due', x= 'state')
```

```python
df_detail.groupby('city', as_index=False).agg({'amt_due':'sum'}).sort_values('amt_due')[-10:].plot.bar(y = 'amt_due', x= 'city')
```

```python
df_detail.groupby('industry', as_index=False).agg({'amt_due':'sum'}).sort_values('amt_due')[-15:].plot.bar(y = 'amt_due', x= 'industry')
```

```python
df_v = df_detail.reset_index()
df_v['year'] = df_v['due_date'].dt.year
df_v['month'] = df_v['due_date'].dt.month
df_v.head()
```

```python
fmt = '${x:,.0f}'
tick = mtick.StrMethodFormatter(fmt)

plot_deposits = df_v.loc[(df_v['year'] == 2019) & (df_v ['deposit'] == True)]
plot_monthly = df_v.loc[(df_v['year'] == 2019) & (df_v ['deposit'] == False)]

plot_deposits = plot_deposits.groupby('city').agg({'customer_id': sum})
plot_monthly = plot_monthly.groupby('city').agg({'customer_id': sum})

fig,axs = plt.subplots (1,2, figsize=(14,6))
axs[0].set_title("2019 Deposit by City")
axs[0].plot(plot_deposits)
axs[0].yaxis.set_major_formatter(tick)

axs[1].set_title("2019 Customer Id Fees")
axs[1].plot(plot_monthly)
axs[1].yaxis.set_major_formatter(tick)

plt.subplots_adjust(wspace = .5)
```
